```python
dic = {}
number = []
data = []



def restart(dictionnary, index) :
    for instance in index : 
        dictionnary[instance] = 0

for sf in sf_total :
    restart(dic, sf_total)
    n = 0
    occurence = []
    for m in methods : #on choisit la méthode
        for i in range(len(m)) : # i correspond à un groupe dans la méthode
            for j in range(len(m[i])) : # j correspond à un élément du groupe
                if sf == m[i][j] : 
                    n += 1
                    for sf2 in m[i] : 
                        dic[sf2] += 1      
    number.append(n) 
    for key, value in dic.items() : 
        occurence.append(value)
    data.append(occurence)
```

```python
Mat = np.asmatrix(data)

pairing_exact = []
titre_exact = []
for i in range(len(sf_total)) : 
    for j in range(len(sf_total)) :
        if number[i] == Mat[j, i] and j < i :
            pairing_exact.append([Mat[j, i], j, i])
            titre_exact.append(str(sf_total[j]) + " ( " + str(number[j]) + " ) - " + str(sf_total[i]) + " ( " + str(number[i]) + " )")
            print(sf_total[j], "(", number[j], ") -", sf_total[i], "(", number[i], ") :", Mat[j, i])
```
